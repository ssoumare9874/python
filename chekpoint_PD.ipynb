```python
#importation des biblietheque
import pandas as pd
import numpy as np
exam_data = {'name' : ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
             'score' : [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],

             'tentatives' : [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
             'qualifier' : ['oui', 'non', 'oui', 'non', 'non', 'oui', 'oui', 'non', 'non', 'oui']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df=pd.DataFrame(exam_data,index=labels)
#formes imprimer
df.head(3)


 
```

```python
#Suppresion des NAN
df=df.dropna()
df
```

```python
#Extraction des colnnes "Name" et "score"
C_S=df[["name","score"]]
C_S
```

```python
#Ajout d'une nouvelle ligne
dat={'name':[ 'Suresh'], 'score' :[ 15.5], 'tentatives' :[ 1], 'qualifier' :['oui']}
df2=pd.DataFrame(dat)
df=pd.concat([df,df2])
df
```

```python
#suppression de la colone attemps
df=df.drop('tentatives',axis=1)
df
```

```python
#Ajoutez une nouvelle colonne 'Succès' : si le score est supérieur à 10, nous aurons 1, sinon 0.
df['succès']=np.where(df['score']>10,1,0)
df
```

```python
#Après avoir exécuté le DataFrame final, exportez-le dans un fichier CSV nommé 'my_data'.


df.to_csv("my_data")
new_def=pd.read_csv("my_data")
new_def
```
